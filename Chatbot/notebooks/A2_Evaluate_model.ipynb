```python
"""
=============================================================================
HỆ THỐNG ĐÁNH GIÁ LLM CHO CHATBOT QUY CHẾ SINH VIÊN
Đánh giá khả năng trả lời và tuân thủ hành vi của mô hình LLM
=============================================================================
Tác giả: Đồ án Chatbot Quy chế Sinh viên
Mô hình: Qwen 2.5 - 7B (Base vs Fine-tuned)
Phương pháp: LLM-only evaluation (không sử dụng RAG)
=============================================================================
"""

# ============================================================================
# BƯỚC 1: CÀI ĐẶT THỦ VIỆN
# ============================================================================

!pip install -q transformers accelerate torch bitsandbytes

import json
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from typing import Dict, List, Tuple
import re
from datetime import datetime
from tqdm import tqdm
import gc

# ============================================================================
# BƯỚC 2: CẤU HÌNH HỆ THỐNG
# ============================================================================

class EvaluationConfig:
    """Cấu hình cho hệ thống đánh giá"""

    # Đường dẫn model
    BASE_MODEL_PATH = "Qwen/Qwen2.5-7B-Instruct"  # Model gốc
    FINETUNED_MODEL_PATH = "wanduc0701/hust_chatbot_qwen2.5_finetuned"  # Model đã fine-tune

    # Đường dẫn dữ liệu
    EVAL_DATA_PATH = "eval_data.json"
    CONTEXT_CORPUS_PATH = "context_corpus.json"

    # Đường dẫn output
    OUTPUT_BASE = "eval_results_base.json"
    OUTPUT_FINETUNED = "eval_results_finetuned.json"

    # Cấu hình inference
    MAX_NEW_TOKENS = 512
    TEMPERATURE = 0.1
    TOP_P = 0.9
    DO_SAMPLE = True

    # Device
    DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# ============================================================================
# BƯỚC 3: TẠO PROMPT TEMPLATE CHO ĐÁNH GIÁ
# ============================================================================

class PromptTemplate:
    """Quản lý prompt template cho đánh giá LLM"""

    @staticmethod
    def create_evaluation_prompt(question: str, context: str) -> str:
        """
        Tạo prompt đánh giá với context được inject

        Args:
            question: Câu hỏi từ sinh viên
            context: Ngữ cảnh từ văn bản quy chế

        Returns:
            Prompt hoàn chỉnh để gửi cho model
        """

        prompt = f"""Bạn là trợ lý tư vấn quy chế sinh viên. Nhiệm vụ của bạn là trả lời câu hỏi dựa HOÀN TOÀN vào nội dung văn bản quy chế được cung cấp.

**QUY TẮC QUAN TRỌNG:**
1. Chỉ trả lời dựa trên thông tin có trong phần "Nội dung quy chế" bên dưới
2. Nếu thông tin không đủ hoặc không liên quan, hãy từ chối lịch sự và giải thích lý do
3. Không bịa đặt hoặc suy đoán thông tin
4. Trả lời ngắn gọn, rõ ràng bằng tiếng Việt

---

**Nội dung quy chế:**
{context}

---

**Câu hỏi từ sinh viên:**
{question}

**Trả lời:**"""

        return prompt

    @staticmethod
    def create_system_message() -> str:
        """Tạo system message cho chat model"""
        return "Bạn là trợ lý tư vấn quy chế sinh viên chuyên nghiệp, chính xác và thân thiện."

# ============================================================================
# BƯỚC 4: LOAD DỮ LIỆU ĐÁNH GIÁ
# ============================================================================

class DataLoader:
    """Load và xử lý dữ liệu đánh giá"""

    @staticmethod
    def load_eval_data(file_path: str) -> List[Dict]:
        """
        Load dữ liệu đánh giá từ file JSON

        Args:
            file_path: Đường dẫn file eval_data.json

        Returns:
            List các mẫu đánh giá
        """
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
            print(f"✅ Đã load {len(data)} mẫu đánh giá từ {file_path}")
            return data
        except FileNotFoundError:
            print(f"❌ Không tìm thấy file: {file_path}")
            return []
        except json.JSONDecodeError as e:
            print(f"❌ Lỗi đọc JSON: {e}")
            return []

    @staticmethod
    def load_context_corpus(file_path: str) -> Dict[str, str]:
        """
        Load corpus ngữ cảnh từ file JSON

        Args:
            file_path: Đường dẫn file context_corpus.json

        Returns:
            Dictionary mapping tên văn bản -> nội dung
        """
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                corpus = json.load(f)
            print(f"✅ Đã load {len(corpus)} văn bản từ {file_path}")
            return corpus
        except FileNotFoundError:
            print(f"❌ Không tìm thấy file: {file_path}")
            return {}
        except json.JSONDecodeError as e:
            print(f"❌ Lỗi đọc JSON: {e}")
            return {}

# ============================================================================
# BƯỚC 5: HỆ THỐNG INFERENCE
# ============================================================================

class LLMEvaluator:
    """Hệ thống inference và đánh giá LLM"""

    def __init__(self, model_path: str, device: str = "cuda"):
        """
        Khởi tạo evaluator với model

        Args:
            model_path: Đường dẫn HuggingFace model
            device: cuda hoặc cpu
        """
        self.model_path = model_path
        self.device = device
        self.model = None
        self.tokenizer = None

        print(f"\n{'='*70}")
        print(f"🔄 Đang load model: {model_path}")
        print(f"{'='*70}")

        self._load_model()

    def _load_model(self):
        """Load model và tokenizer từ HuggingFace"""
        try:
            # Load tokenizer
            self.tokenizer = AutoTokenizer.from_pretrained(
                self.model_path,
                trust_remote_code=True
            )

            # Load model với cấu hình tối ưu
            self.model = AutoModelForCausalLM.from_pretrained(
                self.model_path,
                torch_dtype=torch.float16,
                device_map="auto",
                trust_remote_code=True,
                low_cpu_mem_usage=True
            )

            self.model.eval()
            print(f"✅ Đã load model thành công!")
            print(f"📊 Device: {self.device}")
            print(f"📊 Model size: {sum(p.numel() for p in self.model.parameters()) / 1e9:.2f}B parameters")

        except Exception as e:
            print(f"❌ Lỗi khi load model: {e}")
            raise

    def generate_response(
        self,
        prompt: str,
        max_new_tokens: int = 512,
        temperature: float = 0.1,
        top_p: float = 0.9,
        do_sample: bool = True
    ) -> str:
        """
        Generate response từ model

        Args:
            prompt: Prompt đầu vào
            max_new_tokens: Số token tối đa sinh ra
            temperature: Nhiệt độ sampling
            top_p: Nucleus sampling threshold
            do_sample: Có dùng sampling hay không

        Returns:
            Response text từ model
        """
        try:
            # Tạo messages cho chat model
            messages = [
                {"role": "system", "content": PromptTemplate.create_system_message()},
                {"role": "user", "content": prompt}
            ]

            # Apply chat template
            text = self.tokenizer.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=True
            )

            # Tokenize
            inputs = self.tokenizer(
                text,
                return_tensors="pt",
                truncation=True,
                max_length=4096
            ).to(self.device)

            # Generate
            with torch.no_grad():
                outputs = self.model.generate(
                    **inputs,
                    max_new_tokens=max_new_tokens,
                    temperature=temperature,
                    top_p=top_p,
                    do_sample=do_sample,
                    pad_token_id=self.tokenizer.pad_token_id,
                    eos_token_id=self.tokenizer.eos_token_id
                )

            # Decode response (chỉ lấy phần mới sinh ra)
            response = self.tokenizer.decode(
                outputs[0][inputs['input_ids'].shape[1]:],
                skip_special_tokens=True
            )

            return response.strip()

        except Exception as e:
            print(f"❌ Lỗi khi generate: {e}")
            return "[ERROR: Không thể sinh response]"

    def cleanup(self):
        """Giải phóng bộ nhớ"""
        del self.model
        del self.tokenizer
        gc.collect()
        torch.cuda.empty_cache()
        print("🧹 Đã giải phóng bộ nhớ")

# ============================================================================
# BƯỚC 6: PHÂN TÍCH VÀ ĐÁNH GIÁ HÀNH VI
# ============================================================================

class BehaviorAnalyzer:
    """Phân tích hành vi của model response"""

    # Keywords để detect refusal
    REFUSAL_KEYWORDS = [
        "không có thông tin",
        "không đủ thông tin",
        "không tìm thấy",
        "không rõ",
        "không thể trả lời",
        "xin lỗi",
        "không được cung cấp",
        "không nằm trong",
        "ngoài phạm vi",
        "không liên quan"
    ]

    @staticmethod
    def detect_behavior(response: str) -> str:
        """
        Phát hiện hành vi của model từ response

        Args:
            response: Response text từ model

        Returns:
            'answer' hoặc 'refuse'
        """
        response_lower = response.lower()

        # Kiểm tra có từ khóa từ chối không
        for keyword in BehaviorAnalyzer.REFUSAL_KEYWORDS:
            if keyword in response_lower:
                return "refuse"

        # Nếu response quá ngắn (< 20 ký tự) -> có thể là refuse
        if len(response.strip()) < 20:
            return "refuse"

        # Default: coi như model đã cố gắng trả lời
        return "answer"

    @staticmethod
    def compare_behavior(predicted: str, expected: str) -> bool:
        """
        So sánh hành vi predicted với expected

        Args:
            predicted: Hành vi model thực tế
            expected: Hành vi mong đợi

        Returns:
            True nếu khớp, False nếu không khớp
        """
        return predicted == expected

# ============================================================================
# BƯỚC 7: HỆ THỐNG ĐÁNH GIÁ CHÍNH
# ============================================================================

class MainEvaluator:
    """Orchestrator chính cho toàn bộ quá trình đánh giá"""

    def __init__(self, config: EvaluationConfig):
        """
        Khởi tạo main evaluator

        Args:
            config: Cấu hình đánh giá
        """
        self.config = config
        self.eval_data = []
        self.context_corpus = {}

    def load_data(self):
        """Load tất cả dữ liệu cần thiết"""
        print(f"\n{'='*70}")
        print("📂 ĐANG LOAD DỮ LIỆU...")
        print(f"{'='*70}")

        self.eval_data = DataLoader.load_eval_data(self.config.EVAL_DATA_PATH)
        self.context_corpus = DataLoader.load_context_corpus(self.config.CONTEXT_CORPUS_PATH)

        if not self.eval_data or not self.context_corpus:
            raise ValueError("Không thể load dữ liệu. Vui lòng kiểm tra file!")

    def get_context_for_question(self, expected_doc_type: str) -> str:
        """
        Lấy context phù hợp cho câu hỏi

        Args:
            expected_doc_type: Loại văn bản mong đợi (hoặc null)

        Returns:
            Context text
        """
        if expected_doc_type and expected_doc_type in self.context_corpus:
            return self.context_corpus[expected_doc_type]
        elif expected_doc_type is None:
            # Không có văn bản phù hợp -> trả về context rỗng
            return "[Không có thông tin liên quan trong quy chế]"
        else:
            # Fallback: lấy context từ văn bản đầu tiên
            return list(self.context_corpus.values())[0] if self.context_corpus else ""

    def evaluate_model(self, model_path: str, model_name: str) -> Tuple[List[Dict], Dict]:
        """
        Đánh giá một model

        Args:
            model_path: Đường dẫn model
            model_name: Tên model (để hiển thị)

        Returns:
            Tuple (results_list, metrics_dict)
        """
        print(f"\n{'='*70}")
        print(f"🎯 BẮT ĐẦU ĐÁNH GIÁ: {model_name}")
        print(f"{'='*70}")

        # Khởi tạo evaluator
        evaluator = LLMEvaluator(model_path, self.config.DEVICE)

        results = []
        correct_behaviors = 0
        total_samples = len(self.eval_data)

        # Đánh giá từng mẫu
        for idx, sample in enumerate(tqdm(self.eval_data, desc=f"Evaluating {model_name}")):
            # Lấy context
            context = self.get_context_for_question(sample.get("expected_doc_type"))

            # Tạo prompt
            prompt = PromptTemplate.create_evaluation_prompt(
                sample["question"],
                context
            )

            # Generate response
            response = evaluator.generate_response(
                prompt,
                max_new_tokens=self.config.MAX_NEW_TOKENS,
                temperature=self.config.TEMPERATURE,
                top_p=self.config.TOP_P,
                do_sample=self.config.DO_SAMPLE
            )

            # Phân tích hành vi
            predicted_behavior = BehaviorAnalyzer.detect_behavior(response)
            expected_behavior = sample["expected_behavior"]
            is_correct = BehaviorAnalyzer.compare_behavior(
                predicted_behavior,
                expected_behavior
            )

            if is_correct:
                correct_behaviors += 1

            # Lưu kết quả
            result = {
                "id": sample["id"],
                "question": sample["question"],
                "context_used": context[:200] + "..." if len(context) > 200 else context,
                "expected_behavior": expected_behavior,
                "predicted_behavior": predicted_behavior,
                "behavior_match": is_correct,
                "response": response,
                "ground_truth": sample.get("ground_truth", "")
            }
            results.append(result)

            # In progress mỗi 10 mẫu
            if (idx + 1) % 10 == 0:
                current_accuracy = (correct_behaviors / (idx + 1)) * 100
                print(f"  📊 Progress: {idx + 1}/{total_samples} | Accuracy: {current_accuracy:.2f}%")

        # Tính metrics tổng hợp
        behavior_accuracy = (correct_behaviors / total_samples) * 100 if total_samples > 0 else 0

        # Phân tích chi tiết theo loại hành vi
        answer_samples = [r for r in results if r["expected_behavior"] == "answer"]
        refuse_samples = [r for r in results if r["expected_behavior"] == "refuse"]

        answer_correct = sum(1 for r in answer_samples if r["behavior_match"])
        refuse_correct = sum(1 for r in refuse_samples if r["behavior_match"])

        metrics = {
            "model_name": model_name,
            "model_path": model_path,
            "total_samples": total_samples,
            "behavior_accuracy": behavior_accuracy,
            "correct_behaviors": correct_behaviors,
            "incorrect_behaviors": total_samples - correct_behaviors,
            "answer_samples": len(answer_samples),
            "answer_correct": answer_correct,
            "answer_accuracy": (answer_correct / len(answer_samples) * 100) if answer_samples else 0,
            "refuse_samples": len(refuse_samples),
            "refuse_correct": refuse_correct,
            "refuse_accuracy": (refuse_correct / len(refuse_samples) * 100) if refuse_samples else 0,
            "evaluation_time": datetime.now().isoformat()
        }

        # Cleanup
        evaluator.cleanup()

        return results, metrics

    def save_results(self, results: List[Dict], metrics: Dict, output_path: str):
        """
        Lưu kết quả ra file JSON

        Args:
            results: List kết quả chi tiết
            metrics: Metrics tổng hợp
            output_path: Đường dẫn file output
        """
        output_data = {
            "metrics": metrics,
            "detailed_results": results
        }

        with open(output_path, 'w', encoding='utf-8') as f:
            json.dump(output_data, f, ensure_ascii=False, indent=2)

        print(f"✅ Đã lưu kết quả vào: {output_path}")

    def print_metrics(self, metrics: Dict):
        """In metrics ra màn hình"""
        print(f"\n{'='*70}")
        print(f"📊 KẾT QUẢ ĐÁNH GIÁ: {metrics['model_name']}")
        print(f"{'='*70}")
        print(f"Tổng số mẫu: {metrics['total_samples']}")
        print(f"\n🎯 BEHAVIOR ACCURACY: {metrics['behavior_accuracy']:.2f}%")
        print(f"   ✅ Correct: {metrics['correct_behaviors']}")
        print(f"   ❌ Incorrect: {metrics['incorrect_behaviors']}")
        print(f"\n📋 Chi tiết theo loại:")
        print(f"   • Answer samples: {metrics['answer_samples']}")
        print(f"     → Accuracy: {metrics['answer_accuracy']:.2f}%")
        print(f"   • Refuse samples: {metrics['refuse_samples']}")
        print(f"     → Accuracy: {metrics['refuse_accuracy']:.2f}%")
        print(f"{'='*70}\n")

    def run_full_evaluation(self):
        """Chạy đánh giá đầy đủ cho cả base và fine-tuned model"""
        print(f"\n{'🚀'*35}")
        print("BẮT ĐẦU QUÁ TRÌNH ĐÁNH GIÁ LLM")
        print(f"{'🚀'*35}")

        # Load dữ liệu
        self.load_data()

        # ============= ĐÁNH GIÁ BASE MODEL =============
        print("\n" + "="*70)
        print("📌 GIAI ĐOẠN 1: ĐÁNH GIÁ BASE MODEL")
        print("="*70)

        base_results, base_metrics = self.evaluate_model(
            self.config.BASE_MODEL_PATH,
            "Base Model (Qwen 2.5 - 7B)"
        )

        self.print_metrics(base_metrics)
        self.save_results(base_results, base_metrics, self.config.OUTPUT_BASE)

        # ============= ĐÁNH GIÁ FINE-TUNED MODEL =============
        print("\n" + "="*70)
        print("📌 GIAI ĐOẠN 2: ĐÁNH GIÁ FINE-TUNED MODEL")
        print("="*70)

        finetuned_results, finetuned_metrics = self.evaluate_model(
            self.config.FINETUNED_MODEL_PATH,
            "Fine-tuned Model"
        )

        self.print_metrics(finetuned_metrics)
        self.save_results(finetuned_results, finetuned_metrics, self.config.OUTPUT_FINETUNED)

        # ============= SO SÁNH KẾT QUẢ =============
        self.compare_models(base_metrics, finetuned_metrics)

    def compare_models(self, base_metrics: Dict, finetuned_metrics: Dict):
        """So sánh kết quả giữa base và fine-tuned model"""
        print(f"\n{'='*70}")
        print("📊 SO SÁNH BASE MODEL VS FINE-TUNED MODEL")
        print(f"{'='*70}")

        improvement = finetuned_metrics['behavior_accuracy'] - base_metrics['behavior_accuracy']

        print(f"\n🎯 Behavior Accuracy:")
        print(f"   Base Model:      {base_metrics['behavior_accuracy']:.2f}%")
        print(f"   Fine-tuned:      {finetuned_metrics['behavior_accuracy']:.2f}%")
        print(f"   📈 Improvement:  {improvement:+.2f}%")

        print(f"\n📋 Answer Accuracy:")
        print(f"   Base Model:      {base_metrics['answer_accuracy']:.2f}%")
        print(f"   Fine-tuned:      {finetuned_metrics['answer_accuracy']:.2f}%")
        print(f"   📈 Improvement:  {finetuned_metrics['answer_accuracy'] - base_metrics['answer_accuracy']:+.2f}%")

        print(f"\n🚫 Refuse Accuracy:")
        print(f"   Base Model:      {base_metrics['refuse_accuracy']:.2f}%")
        print(f"   Fine-tuned:      {finetuned_metrics['refuse_accuracy']:.2f}%")
        print(f"   📈 Improvement:  {finetuned_metrics['refuse_accuracy'] - base_metrics['refuse_accuracy']:+.2f}%")

        print(f"\n{'='*70}")

        if improvement > 0:
            print("✅ Fine-tuning đã cải thiện performance!")
        elif improvement < 0:
            print("⚠️ Fine-tuning làm giảm performance. Cần xem xét lại dữ liệu training.")
        else:
            print("➡️ Không có sự thay đổi đáng kể.")

        print(f"{'='*70}\n")

# ============================================================================
# BƯỚC 8: MAIN EXECUTION
# ============================================================================

def main():
    """Hàm main để chạy toàn bộ pipeline đánh giá"""

    print("""
    ╔══════════════════════════════════════════════════════════════════╗
    ║                                                                  ║
    ║     HỆ THỐNG ĐÁNH GIÁ LLM - CHATBOT QUY CHẾ SINH VIÊN          ║
    ║                                                                  ║
    ║     Model: Qwen 2.5 - 7B Instruct                               ║
    ║     Method: Behavior-based Evaluation (LLM-only)                ║
    ║                                                                  ║
    ╚══════════════════════════════════════════════════════════════════╝
    """)

    # Khởi tạo cấu hình
    config = EvaluationConfig()

    # Tạo evaluator
    evaluator = MainEvaluator(config)

    # Chạy đánh giá đầy đủ
    try:
        evaluator.run_full_evaluation()

        print("\n" + "="*70)
        print("✅ HOÀN THÀNH ĐÁNH GIÁ!")
        print("="*70)
        print(f"📁 Kết quả đã được lưu tại:")
        print(f"   • Base Model: {config.OUTPUT_BASE}")
        print(f"   • Fine-tuned Model: {config.OUTPUT_FINETUNED}")
        print("="*70 + "\n")

    except Exception as e:
        print(f"\n❌ LỖI XẢY RA: {e}")
        import traceback
        traceback.print_exc()

# ============================================================================
# CHẠY CHƯƠNG TRÌNH
# ============================================================================

if __name__ == "__main__":
    main()
```

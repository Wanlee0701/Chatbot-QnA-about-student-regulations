- **ƒê·∫ßu ra (Output):** C√°c file `.txt` ch·ª©a n·ªôi dung vƒÉn b·∫£n thu·∫ßn t√∫y ƒë∆∞·ª£c tr√≠ch xu·∫•t t·ª´ c√°c file PDF.

**üìù L∆∞u √Ω quan tr·ªçng:**
- Phi√™n b·∫£n n√†y **kh√¥ng t·ª± ƒë·ªông x·ª≠ l√Ω b·∫£ng bi·ªÉu**. C·∫•u tr√∫c c·ªßa b·∫£ng s·∫Ω b·ªã m·∫•t v√† ƒë∆∞·ª£c tr√≠ch xu·∫•t d∆∞·ªõi d·∫°ng vƒÉn b·∫£n th√¥. B·∫°n s·∫Ω c·∫ßn ph·∫£i x·ª≠ l√Ω ho·∫∑c ƒë·ªãnh d·∫°ng l·∫°i c√°c ph·∫ßn ch·ª©a b·∫£ng bi·ªÉu n√†y m·ªôt c√°ch **th·ªß c√¥ng** trong c√°c file `.txt` ƒë·∫ßu ra.

- ∆Øu ƒëi·ªÉm l·ªõn l√† t·ªëc ƒë·ªô x·ª≠ l√Ω r·∫•t nhanh v√† y√™u c·∫ßu √≠t th∆∞ vi·ªán ph·ª• thu·ªôc h∆°n.


## 1. Chu·∫©n B·ªã M√¥i Tr∆∞·ªùng
Ph∆∞∆°ng ph√°p n√†y ch·ªâ y√™u c·∫ßu m·ªôt th∆∞ vi·ªán ch√≠nh l√† `pypdf` v√† `tqdm` ƒë·ªÉ hi·ªÉn th·ªã thanh ti·∫øn tr√¨nh.

```python
import os
from pypdf import PdfReader
from tqdm import tqdm # Th√™m tqdm ƒë·ªÉ theo d√µi ti·∫øn tr√¨nh
```

## 2. H√†m l√µi: Chuy·ªÉn ƒë·ªïi m·ªôt file PDF

ƒê√¢y l√† h√†m ch√≠nh, nh·∫≠n v√†o ƒë∆∞·ªùng d·∫´n c·ªßa m·ªôt file PDF v√† ƒë∆∞·ªùng d·∫´n file output. N√≥ s·ª≠ d·ª•ng th∆∞ vi·ªán `pypdf` ƒë·ªÉ:
1. M·ªü file PDF.
2. L·∫∑p qua t·ª´ng trang.
3. Tr√≠ch xu·∫•t to√†n b·ªô vƒÉn b·∫£n t·ª´ m·ªói trang.
4. N·ªëi t·∫•t c·∫£ vƒÉn b·∫£n l·∫°i v√† l∆∞u v√†o m·ªôt file `.txt` duy nh·∫•t.


```python
def convert_pdf_to_txt(pdf_path, output_path):
    """
    S·ª≠ d·ª•ng pypdf ƒë·ªÉ tr√≠ch xu·∫•t to√†n b·ªô n·ªôi dung vƒÉn b·∫£n t·ª´ m·ªôt file PDF.
    """
    print(f"üìÑ ƒêang x·ª≠ l√Ω file: {os.path.basename(pdf_path)}...")
    try:
        reader = PdfReader(pdf_path)
        full_text = ""
        # L·∫∑p qua t·ª´ng trang v√† l·∫•y text
        for page in reader.pages:
            full_text += page.extract_text() + "\n\n" # Th√™m hai d√≤ng m·ªõi ƒë·ªÉ ph√¢n t√°ch c√°c trang
        
        # Ghi n·ªôi dung ƒë√£ tr√≠ch xu·∫•t v√†o file TXT
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(full_text)
        # print(f"‚úÖ X·ª≠ l√Ω th√†nh c√¥ng, ƒë√£ l∆∞u v√†o: {output_path}") # C√≥ th·ªÉ b·ªè qua ƒë·ªÉ output g·ªçn h∆°n khi ch·∫°y h√†ng lo·∫°t

    except Exception as e:
        print(f"‚ùå L·ªói khi x·ª≠ l√Ω file {pdf_path}: {e}")
```

## 3. H√†m x·ª≠ l√Ω: Qu√©t v√† ch·∫°y tr√™n to√†n b·ªô th∆∞ m·ª•c

H√†m n√†y s·∫Ω t·ª± ƒë·ªông t√¨m t·∫•t c·∫£ c√°c file c√≥ ƒëu√¥i `.pdf` trong th∆∞ m·ª•c ngu·ªìn (`dataset`) v√† g·ªçi h√†m `convert_pdf_to_txt` cho t·ª´ng file.


```python
def process_all_pdfs_in_folder(data_raw, output_folder):
    """
    L·∫∑p qua t·∫•t c·∫£ c√°c file PDF trong th∆∞ m·ª•c ngu·ªìn v√† x·ª≠ l√Ω ch√∫ng.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

        # L·∫•y danh s√°ch c√°c file PDF
    pdf_files = [f for f in os.listdir(data_raw) if f.lower().endswith(".pdf")]

    if not pdf_files:
        print(f"Kh√¥ng t√¨m th·∫•y file PDF n√†o trong th∆∞ m·ª•c '{data_raw}'.")
        return

    # S·ª≠ d·ª•ng tqdm ƒë·ªÉ t·∫°o thanh ti·∫øn tr√¨nh
    for filename in tqdm(pdf_files, desc="T·ªïng ti·∫øn tr√¨nh x·ª≠ l√Ω PDF"):
        pdf_path = os.path.join(data_raw, filename)
        output_filename = os.path.splitext(filename)[0] + ".txt"
        output_path = os.path.join(output_folder, output_filename)
        
        convert_pdf_to_txt(pdf_path, output_path)
```

## üöÄ 4. Ch·∫°y ch∆∞∆°ng tr√¨nh!


```python
# --- ƒê·ªãnh nghƒ©a th∆∞ m·ª•c ngu·ªìn v√† ƒë√≠ch ---
PDF_SOURCE_FOLDER = os.path.join("..", "data", "data_raw") 

# T∆∞∆°ng t·ª±, ƒëi l√™n th∆∞ m·ª•c g·ªëc r·ªìi v√†o 'data/processed'
TXT_OUTPUT_FOLDER = os.path.join("..", "data", "processed")


# --- B·∫Øt ƒë·∫ßu th·ª±c thi ---
if not os.path.isdir(PDF_SOURCE_FOLDER):
    print(f"L·ªói: Th∆∞ m·ª•c ngu·ªìn '{PDF_SOURCE_FOLDER}' kh√¥ng t·ªìn t·∫°i. Vui l√≤ng ki·ªÉm tra l·∫°i c·∫•u tr√∫c th∆∞ m·ª•c.")
else:
    print(f"B·∫Øt ƒë·∫ßu qu√° tr√¨nh chuy·ªÉn ƒë·ªïi h√†ng lo·∫°t t·ª´ '{PDF_SOURCE_FOLDER}'...")
    process_all_pdfs_in_folder(PDF_SOURCE_FOLDER, TXT_OUTPUT_FOLDER)
    print(f"\nüéâ Ho√†n t·∫•t! K·∫øt qu·∫£ ƒë√£ ƒë∆∞·ª£c l∆∞u trong th∆∞ m·ª•c '{TXT_OUTPUT_FOLDER}'.")
```
